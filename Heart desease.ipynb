```python
def drop_outliers(hs, field_name):
    iqr = 1.5 * (np.percentile(hs[field_name],70) - np.percentile(hs[field_name], 0))
    hs.drop(hs[hs[field_name] > (iqr + np.percentile(hs[field_name],30))].index, inplace=True)
    hs.drop(hs[hs[field_name] < (np.percentile(hs[field_name], 0) - iqr)].index, inplace=True)
    drop_outliers(hs ,"Chest pain type")
    drop_outliers(hs ,"BP")
    drop_outliers(hs ,"Cholesterol")
    drop_outliers(hs ,"FBS over 120")
    drop_outliers(hs ,"Max HR")
    drop_outliers(hs ,"ST depression")
    drop_outliers(hs ,"Number of vessels fluro")
    drop_outliers(hs ,"Age")
    drop_outliers(hs ,"EKG results")
    drop_outliers(hs ,"Exercise angina")
    drop_outliers(hs ,"Slope of ST")
    drop_outliers(hs ,"Thallium")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
hs ["Heart Disease"] = le.fit_transform(hs["Heart Disease"])
hs["Heart Disease"]
```

```python
hs
```

```python
y =hs ["Heart Disease"]
x = hs.drop("Heart Disease", axis=1)
```

```python
y
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2) 
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,  y_train)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = reg.predict(x_test)

print("MEA", mean_absolute_error(y_test,y_pred))
```

```python
r2_score= reg.score(x_test, y_test)

print("R^2 score: ", r2_score)
print("predicted target values:", y_pred)
```

```python
plt.scatter(y_pred, y_test)
```

```python
plt.plot(y_test, label='True Values')
plt.plot(y_pred, label='Predicted Values')
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg= DecisionTreeRegressor()
dt_reg.fit(x_train, y_train)
y_pred= dt_reg.predict(x_test)
 
print("MEA", mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.svm import SVR

sv_reg = SVR()
sv_reg.fit(x_train,y_train)
sv_pred=sv_reg.predict(x_test)
print("MEA", mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print("MEA", mean_absolute_error(y_test,y_pred))
```
