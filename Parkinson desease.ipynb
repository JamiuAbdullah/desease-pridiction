```python
def drop_outliers(pk, field_name):
    iqr = 1.5 * (np.percentile(pk[field_name],75) - np.percentile(pk[field_name], 25))
    pk.drop(pk[pk[field_name] > (iqr + np.percentile(pk[field_name],30))].index, inplace=True)
    pk.drop(pk[pk[field_name] < (np.percentile(pk[field_name], 25) - iqr)].index, inplace=True)
    
    drop_outliers(pk, 'MDVP:Fo(Hz)') 
    drop_outliers(pk, 'MDVP:Fhi(Hz)') 
    drop_outliers(pk, 'MDVP:Flo(Hz)') 
    drop_outliers(pk, 'MDVP:Jitter(%)') 
    drop_outliers(pk, 'MDVP:Jitter(Abs)') 
    drop_outliers(pk, 'MDVP:RAP') 
    drop_outliers(pk, 'MDVP:PPQ') 
    drop_outliers(pk, 'Jitter:DDP') 
    drop_outliers(pk, 'MDVP:Shimmer') 
    drop_outliers(pk, 'MDVP:Shimmer(dB)') 
    drop_outliers(pk, 'Shimmer:APQ3') 
    drop_outliers(pk, 'Shimmer:APQ5') 
    drop_outliers(pk, 'MDVP:APQ') 
    drop_outliers(pk, 'Shimmer:DDA') 
    drop_outliers(pk, 'NHR') 
    drop_outliers(pk, 'HNR') 
    drop_outliers(pk, 'status') 
    drop_outliers(pk, 'RPDE') 
    drop_outliers(pk, 'DFA')  
    drop_outliers(pk, 'D2') 
    drop_outliers(pk, 'PPE') 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
reg.fit(x_train, y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
accuracy_training_set = accuracy_score(y_test, y_pred)
f1_score_set = f1_score(y_test,y_pred) # type: ignore
precision_score_set = precision_score(y_test,y_pred)
recall_score_set = recall_score(y_test,y_pred)
```
